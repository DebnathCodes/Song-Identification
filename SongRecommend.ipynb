```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io



file_name = next(iter(uploaded))  # Get the filename from the uploaded dictionary
df = pd.read_csv(io.BytesIO(uploaded[file_name])) # Read the file into a pandas DataFrame
```

```python

import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# Drop unnecessary columns
df = df.drop(columns=['link'], errors='ignore')  # Ignore error if column doesn't exist

# Create a combined text column (song name, artist, and lyrics)
df['full_text'] = df['song'] + " by " + df['artist'] + " - " + df['text']

# Initialize the TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)

# Fit and transform the dataset's full text
tfidf_matrix = vectorizer.fit_transform(df['full_text'])

# Function to identify the closest matching song
def identify_song_tfidf(query):
    # Transform the query using the same TF-IDF vectorizer
    query_vector = vectorizer.transform([query])

    # Compute cosine similarity between query and dataset
    similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()

    # Get the best matching song
    best_match_index = np.argmax(similarities)
    best_match = df.iloc[best_match_index]

    return best_match[['song', 'artist']]

# Take user input dynamically
query_text = input("Enter a snippet of the song lyrics: ")

# Identify the song
identified_song = identify_song_tfidf(query_text)

# Print the identified song
print("\n Identified Song:", identified_song['song'])
print("Artist:", identified_song['artist'])
```
